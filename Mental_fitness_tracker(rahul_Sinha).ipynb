```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/share-with-mental-and-substance-disorders2.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype == 'object':
    df[i]=l.fit_transform(df[i])
```

```python
df.shape
```

Split Data

```python
x = df.drop('Mental fitness',axis='columns')
y = df['Mental fitness']

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.20, random_state=2)
```

```python
print("xtrain: ", xtrain.shape)
print("xtest: ", xtest.shape)
print("\n ytrain: ", ytrain.shape)
print("ytest: ", ytest.shape)
```

Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(xtrain, ytrain)


ytrain_pred = lr.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))

r2 = r2_score(ytrain, ytrain_pred)

print("The Linear Regression model performence for training set")
print("---------------------------------")
print('MSE is {}', format(mse))
print('RMSE is {}', format(rmse))
print('R2 score is {}', format(r2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)

ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)

print("The Random Forest Regressor model performence for training set")
print("------------------------")
print('MSE is {}', format(mse))
print('RMSE is {}', format(rmse))
print('R2 score is {}', format(r2))
```

```python
ytest_pred = lr.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))

r2 = r2_score(ytest, ytest_pred)

print("The Linear Regression model performence for training set")
print("---------------------------------")
print('MSE is {}', format(mse))
print('RMSE is {}', format(rmse))
print('R2 score is {}', format(r2))
print("---------------------------------")

ytest_pred = rf.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)

print("The Random Forest Regressor model performence for training set")
print("---------------------------------")
print('MSE is {}', format(mse))
print('RMSE is {}', format(rmse))
print('R2 score is {}', format(r2))
```
